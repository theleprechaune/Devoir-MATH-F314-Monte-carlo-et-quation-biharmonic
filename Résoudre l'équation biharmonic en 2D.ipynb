# Discretisation de l'équation biharmonic 

Nous avons l'équation biharmonic : 

\begin{equation}
\nabla^4 \psi= 0
\end{equation}

ceci nous donne en 2D : 

\begin{equation}
\frac{\partial^4 \psi}{\partial x^4} + \frac{\partial^4 \psi}{\partial y^4} + \frac{\partial^4 \psi}{\partial x^2 \partial y^2} = 0
\end{equation}

nous utilisons cette page wikipidia : https://en.wikipedia.org/wiki/Finite_difference_coefficient pour la discretiser en utilisant la méthode des différences finie central avec une précision du 2ème ordre. Nous obtenons : 

\begin{equation}
\frac{\partial^4 \psi_{i,j}}{\partial x^4} = \frac{\psi_{i-2,j} - 4 \psi_{i-1,j} + 6 \psi_{i,j} - 4 \psi_{i+1,j} + \psi_{i+2,j}}{\Delta x^4}
\end{equation}

\begin{equation}
\frac{\partial^4 \psi_{i,j}}{\partial y^4} = \frac{\psi_{i,j-2} - 4 \psi_{i,j-1} + 6 \psi_{i,j} - 4 \psi_{i,j+1} + \psi_{i,j+2}}{\Delta y^4}
\end{equation}

\begin{equation}
\frac{\partial^4 \psi_{i,j}}{\partial x^2 \partial y^2} = \frac{ 2 \psi_{i-1,j-1} -4 \psi_{i-1,j} + 2\psi_{i-1,j+1}
                                                                 - 4 \psi_{i,j-1} +8 \psi_{i,j} -4 \psi_{i,j+1}
                                                                 + 2 \psi_{i+1,j-1} -4 \psi_{i+1,j} + 2\psi_{i+1,j+1} 
}{\Delta x^2 \Delta y^2}
\end{equation}

En choisissant $\Delta x = \Delta y$ , nous obtenons l'expression suivant pour l'opérateur biharmonic en 2D :

\begin{equation}
\nabla^4 \psi= \frac{ 20 \psi_{i,j}
                      - 8 \psi_{i-1,j} - 8 \psi_{i+1,j} - 8 \psi_{i,j-1} - 8 \psi_{i,j+1}
                      + \psi_{i-2,j} + \psi_{i+2,j} + \psi_{i,j-2} + \psi_{i,j+2}
                      + 2\psi_{i-1,j-1} + 2\psi_{i-1,j+1} + 2\psi_{i+1,j-1} + 2\psi_{i+1,j+1}}{\Delta x^4}
\end{equation}

Les points $\psi_{0,j},\psi_{1,j},\psi_{nx,j},\psi_{nx-1,j},\psi_{i,0},\psi_{i,1},\psi_{i,ny},\psi_{i,ny-1}$ ne peuvent pas être calculer avec cette méthode car elle demande des points qui sont en dehors de notre domaine . Nous calculerons ces points grâce au condition aux bords.

# Discretisation des conditions aux bords

Nous avons les conditions suivante pour le bord gauche :

\begin{equation}
\frac{\partial \psi}{\partial x}= 0
\end{equation}

et 

\begin{equation}
\psi = 0
\end{equation}

Elles sont justifiées par le fait que le bord gauche est un mur, le flux est donc nulle en ces points(même chose pour le bord droit et bas).
En utilisant la méthode des différence finie avant avec une précision du 2e ordre pour la dérivé partielle en x, nous obtenons la version discretisée suivante  : 

\begin{equation}
\psi_{0,j}= 0
\end{equation}

et

\begin{equation}
\frac{ -3\psi_{0,j} +4\psi_{1,j} -\psi_{2,j}}{\Delta x}= 0 \\
\psi_{1,j} = \frac{\psi_{2,j}}{4}
\end{equation}


Nous avons les conditions suivante pour le bord droit :

\begin{equation}
\frac{\partial \psi}{\partial x}= 0
\end{equation}

et 

\begin{equation}
\psi = 0
\end{equation}

En utilisant la méthode des différence finie arrière avec une précision du 2e ordre pour la dérivé partielle en x, nous obtenons la version discretisée suivante  : 

\begin{equation}
\psi_{nx,j}= 0
\end{equation}

et

\begin{equation}
\frac{ \psi_{nx-2,j} -4\psi_{nx-1,j} +3\psi_{nx,j}}{\Delta x}= 0 \\
\psi_{nx-1,j} = \frac{\psi_{nx-2,j}}{4}
\end{equation}


Nous avons les conditions suivante pour le bord du bas :

\begin{equation}
\frac{\partial \psi}{\partial y}= 0
\end{equation}

et 

\begin{equation}
\psi = 0
\end{equation}

En utilisant la méthode des différence finie avant avec une précision du 2e ordre pour la dérivé partielle en y, nous obtenons la version discretisée suivante  : 

\begin{equation}
\psi_{i,0}= 0
\end{equation}

et

\begin{equation}
\frac{ -3\psi_{i,0} +4\psi_{i,1} -\psi_{i,2}}{\Delta y}= 0 \\
\psi_{i,1} = \frac{\psi_{i,2}}{4}
\end{equation}

Nous avons les conditions suivante pour le bord du haut :

\begin{equation}
\frac{\partial \psi}{\partial y}= 1
\end{equation}

et 

\begin{equation}
\psi = 0
\end{equation}

elles sont justifiées par le fait que le bord du haut agit comme un mur qui bouge à vitesse constante mais qui empèche le liquide de s'échapper donc $\frac{\partial \psi}{\partial y}= u = 1 $ où u est la vitesse du fluide selon l'axe x et $\psi = 0$ tout le long du mur implique $\frac{\partial \psi}{\partial x}= -v = 0 $ où v est la vitesse selon l'axe y, ainsi, le liquide ne peut pas passer le mur.
En utilisant la méthode des différence finie arrière avec une précision du 2e ordre pour la dérivé partielle en y, nous obtenons la version discretisée suivante  : 

\begin{equation}
\psi_{i,ny}= 0
\end{equation}

et

\begin{equation}
\frac{ \psi_{i,ny-2} -4\psi_{i,ny-1} +3\psi_{i,ny}}{\Delta y}= 1 \\
\psi_{i,ny-1} = \frac{\psi_{i,ny-2} - \Delta y}{4}
\end{equation}





# Implementation de la méthode du gradient conjugé

Nous pouvons copier le code dévlopper dans le cours. Il suffit alors de changer l'opérateur laplacien par l'opérateur biharmonic 2D calculé et d'y ajouter les conditions de Neuman.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
nx = 41
ny = 41

l = 1.
h = 1.

dx = l/(nx-1)
dy = h/(ny-1)

l1_target = 1e-8
```

```python
def L1norm(new, old):
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

```python
def bihar_2d_conjugate_gradient(p0, dx, dy,
                                  maxiter=20000, rtol=1e-8):
    """
    Solves the 2D Biharmonic equation on a uniform grid,
    with the same grid spacing in both directions,
    using the method of conjugate gradients.
    
    The function assumes Dirichlet boundary conditions with value zero.
    The exit criterion of the solver is based on the relative L1norm
    of the solution difference between two consecutive iterations.

    Parameters
    ----------
    p0 : numpy.ndarray
        The initial solution as a 2D array of floats.
    dx : float
        Grid spacing in the x direction.
    dy : float
        Grid spacing in the y direction.
    maxiter : integer, optional
        Maximum number of iterations to perform;
        default: 20000.
    rtol : float, optional
        Relative tolerance for convergence;
        default: 1e-6.

    Returns
    -------
    p : numpy.ndarray
        The solution after relaxation as a 2D array of floats.
    ite : integer
        The number of iterations performed.
    conv : list
        The convergence history as a list of floats.
    """
    def A(p):
        # Apply the biharmonic operator to p.
        return ( 20.0 * p[2:-2, 2:-2] +
                p[0:-4, 2:-2] + p[4: , 2:-2] + p[2:-2, :-4] + p[2:-2, 4: ]
                - 8 * p[1:-3, 2:-2] - 8 * p[3:-1, 2:-2] -  8 * p[2:-2, 1:-3] - 8 * p[2:-2, 3:-1]
               +  2 * p[1:-3, 1:-3] + 2 * p[1:-3, 3:-1] + 2 * p[3:-1, 1:-3] + 2 * p[3:-1, 3:-1]
               ) / dx**4
    p = p0.copy()
    r = numpy.zeros_like(p)  # initial residual
    Ad = numpy.zeros_like(p)  # to store the mat-vec multiplication
    conv = []  # convergence history
    diff = rtol + 1  # initial difference
    ite = 0  # iteration index
    # Compute the initial residual.
    r[2:-2, 2:-2] = - A(p)
    # Set the initial search direction to be the residual.
    d = r.copy()
    while diff > rtol and ite < maxiter:
        pk = p.copy()
        rk = r.copy()
        # Compute the Biharmonic of the search direction.
        Ad[2:-2, 2:-2] = A(d)
        # Compute the step size.
        alpha = numpy.sum(r * r) / numpy.sum(d * Ad)
        # Update the solution.
        p = pk + alpha * d
        # Update the residual.
        r = rk - alpha * Ad
        # Update the search direction.
        beta = numpy.sum(r * r) / numpy.sum(rk * rk)
        d = r + beta * d
        # Dirichlet boundary conditions are automatically enforced.
        #left neuman boundary
        p[1:-1, 1] = p[1:-1, 2]/4
        # right neuman boundary
        p[1:-1, -2] = p[1:-1, -3]/4
        # bottom neuman boundary
        p[1, 1:-1] = p[2, 1:-1]/4
        # top neuman boundary
        p[-2, 1:-1] = ( p[-3, 1:-1] - dy ) / 4 
        # Compute the relative L1norm of the difference.
        diff = L1norm(p, pk)
        conv.append(diff)
        ite += 1
    return p, ite, conv
```

```python
p0 = numpy.zeros((nx,ny))
p0[-2 , 1:-1] = -dy/4 # Our first guess of the solution taking into account the top neuman boundary
```

```python
solution,iteration,history = bihar_2d_conjugate_gradient(p0, dx, dy,20000,1e-8)
```

```python
from mpl_toolkits import mplot3d
from matplotlib import cm
```

```python
def plot_3d(x, y, p, label='$z$', elev=30.0, azim=45.0):
    """
    Creates a Matplotlib figure with a 3D surface plot
    of the scalar field p.

    Parameters
    ----------
    x : numpy.ndarray
        Gridline locations in the x direction as a 1D array of floats.
    y : numpy.ndarray
        Gridline locations in the y direction as a 1D array of floats.
    p : numpy.ndarray
        Scalar field to plot as a 2D array of floats.
    label : string, optional
        Axis label to use in the third direction;
        default: 'z'.
    elev : float, optional
        Elevation angle in the z plane;
        default: 30.0.
    azim : float, optional
        Azimuth angle in the x,y plane;
        default: 45.0.
    """
    fig = pyplot.figure(figsize=(8.0, 6.0))
    ax = mplot3d.Axes3D(fig)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel(label)
    X, Y = numpy.meshgrid(x, y)
    ax.plot_surface(X, Y, p, cmap=cm.viridis)
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(y[0], y[-1])
    ax.view_init(elev=elev, azim=azim)
```

# Valeur de la solution

Nous allons trouver des valeur particulière de la solution. Pour cela, il suffit de créer une function qui convertit un point de l'espace en index

```python
def find_index(x,y,dx,dy): 
    """
    x : x value of the grid, can be float or integer
    y : y value of the grid, can be float or integer
    dx : step in the x direction of the grid, flot or integer
    dy : step in the y direction of the grid, flot or integer
    ---------
    indx : index of the x number, integer
    indy : index of the y number, integer
    
    """
    indx = round(x/dx)
    indy = round(y/dy)
    return indx, indy
```
